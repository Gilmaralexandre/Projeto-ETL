```python
#importação das bibliotecas
import pandas as pd 
import pandera as pa
```

```python
#utilizando o pandas para ler o csv, e visualizar o dataframe
#parse_dates para tratar datas
df = pd.read_csv('ocorrencia.csv', sep = ";",parse_dates=['ocorrencia_dia'], dayfirst=True)
df
```

```python
#propriedade que mostra os tipo de dados
df.dtypes
```

```python
#com a importação do pandera, da pra utilizar o metodo dataframeshema para definir a coluna como data
#caso aconteça alguma inserção de dado diferente, a continuação do codigo é cessado, e o erro é informado
#coluna ocorrencia dia foi utilizado uma expressão regular('pesquisar depois')
schema = pa.DataFrameSchema(
    columns = {
        "codigo_ocorrencia":pa.Column(pa.Int),
        "codigo_ocorrencia2":pa.Column(pa.Int),
        "ocorrencia_classificacao":pa.Column(pa.String),
        "ocorrencia_cidade":pa.Column(pa.String),
        "ocorrencia_uf":pa.Column(pa.String),
        "ocorrencia_aerodromo":pa.Column(pa.String),
        "ocorrencia_dia":pa.Column(pa.DateTime),
        "ocorrencia_hora":pa.Column(pa.String, pa.Check.str_matches(r'^([0-1]?[0-9]|[2][0-3]):([0-5][0-9])(:[0-5][0-9])?$'), nullable=True),
        "total_recomendacoes":pa.Column(pa.Int)
    })
```

```python
schema.validate(df)
```
